```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('Outputs/neural_net_perf.txt', 'r') as f:
    perf = f.readlines()
```

```python
train = [p for p in perf if 'Epoch' in p]
test = [p for p in perf if 'Test' in p]
```

```python
train
```

```python
train_loss = [float(p.split(' = ')[1].split(';')[0]) for p in train]
train_accuracy = [float(p.split('Accuracy = ')[1].split('; ')[0]) for p in train]
train_lr = [float(p.split('LR = ')[1].split('\n')[0]) for p in train]
```

```python
test_loss = [float(p.split(' Loss: ')[1].split(';')[0]) for p in test]
test_accuracy = [float(p.split(' Accuracy: ')[1].split('\n')[0]) for p in test]
```

```python
sns.lineplot(x=list(range(1, 1+len(train_loss))), y=train_loss, color='blue', label='Training loss')
sns.lineplot(x=list(range(1, 1+len(train_loss), 10)), y=test_loss, color='red', label='Test loss')
plt.title('Cross Entropy Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Cross Entropy Loss')
plt.show()
```

```python
sns.lineplot(x=list(range(1, 1+len(train_loss))), y=train_accuracy, color='blue', label='Training accuarcy')
sns.lineplot(x=list(range(1, 1+len(train_loss), 10)), y=test_accuracy, color='red', label='Test accuracy')
plt.title('Accuracy vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```

```python
sns.lineplot(x=list(range(1, 1+len(train_loss))), y=train_lr, color='blue')
plt.title('Learning Rate vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.show()
```
